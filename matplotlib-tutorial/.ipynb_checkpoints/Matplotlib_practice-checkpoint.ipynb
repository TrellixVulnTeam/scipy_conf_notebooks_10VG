```python
#import basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S = np.cos(x),np.sin(x)
```

```python
np.linspace ## By default it returns list of length 50 if count not specified
```

```python
plt.plot(x,C,color='blue',linewidth=2.0,linestyle='-')
plt.plot(x,S,color='green',linewidth=2.0,linestyle='-')
plt.plot(C,S,color='red')
#plt.plot(S,C)
#set x limits
plt.xlim(-4.0,4.0)
#set x ticks
plt.xticks(np.linspace(-4.,4.,9,endpoint=True))
#set y limit
plt.ylim(-1.0,1.0)
#set y ticks
plt.yticks(np.linspace(-1.0,1.0,5,endpoint=True))
##
```

```python
plt.plot(x,C,color='blue',linewidth=2.0,linestyle='-',label='cosine')
plt.plot(x,S,color='green',linewidth=2.0,linestyle='-',label='sine')
plt.plot(C,S,color='red',label='Sin-Cos')
plt.legend(loc='upper left')
#plt.plot(S,C)
#set x limits
plt.xlim(-4.0,4.0)
#set x ticks
plt.xticks(np.linspace(-4.,4.,9,endpoint=True))
#set y limit
plt.ylim(-1.0,1.0)
#set y ticks
plt.yticks(np.linspace(-1.0,1.0,5,endpoint=True))
##ax
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n=256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot(X,Y+1,color='blue',alpha=1.00)
plt.fill_between(X,1,Y+1,color='blue')
plt.plot(X,Y-1,color='blue',alpha=1.00)
plt.fill_between(X,-1,Y-1,where=(Y<0),color='red')
plt.fill_between(X,-1,Y-1,where=(Y>0),color='green',alpha=25)
```

```python
## import numpy as np
import matplotlib.pyplot as plt

n=12
X=np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X,Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+.04,y+.02,'%2.2f'%y,ha='center',va='bottom')

for x,y in zip(X,Y2):
    plt.text(x-.04,-y-.15,'%2.2f'%y,ha='center',va='bottom')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap='jet')
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)
```
